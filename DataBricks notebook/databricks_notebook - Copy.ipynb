```python
# monthly_sales_trend

from pyspark.sql.functions import sum, avg, count, col, month, year, dense_rank, row_number
from pyspark.sql.window import Window

# 1. Monthly Sales Trend
df_monthly_sales = df_retails_silver.withColumn("year", year(col("transaction_date"))) \
    .withColumn("month", month(col("transaction_date"))) \
    .groupBy("year", "month") \
    .agg(
        sum(col("total_amount")).alias("total_revenue"),
        count(col("transaction_id")).alias("total_transactions"),
        avg(col("total_amount")).alias("avg_transaction_value")
    ) \
    .orderBy("year", "month")

df_monthly_sales.write.mode('overwrite').saveAsTable('--catalogname--.--schema name--.monthly_sales_trend_gold')
```

```python
# Top Customers by Revenue
df_top_customers = df_retails_silver.groupBy("customer_id", "full_name", "email", "country") \
    .agg(
        sum(col("total_amount")).alias("total_spent"),
        count(col("transaction_id")).alias("total_purchases"),
        avg(col("total_amount")).alias("avg_purchase_value")
    ) \
    .orderBy(col("total_spent").desc()) \
    .limit(10)

df_top_customers.write.mode('overwrite').saveAsTable('--catalogname--.--schema name--.top_customers_gold')
```

```python
# Store Performance Analysis
df_store_performance = df_retails_silver.groupBy("store_id", "store_name", "location") \
    .agg(
        sum(col("total_amount")).alias("total_revenue"),
        count(col("transaction_id")).alias("total_transactions"),
        avg(col("total_amount")).alias("avg_transaction_value"),
        sum(col("quantity")).alias("total_items_sold")
    ) \
    .orderBy(col("total_revenue").desc())

df_store_performance.write.mode('overwrite').saveAsTable('--catalogname--.--schema name--.store_performance_gold')
```

```python
# Category Performance Analysis
df_category_performance = df_retails_silver.groupBy("category") \
    .agg(
        sum(col("total_amount")).alias("total_revenue"),
        count(col("transaction_id")).alias("total_transactions"),
        sum(col("quantity")).alias("total_quantity_sold"),
        avg(col("price")).alias("avg_product_price")
    ) \
    .orderBy(col("total_revenue").desc())

df_category_performance.write.mode('overwrite').saveAsTable('--catalogname--.--schema name--.category_performance_gold')
```

```python
from pyspark.sql.functions import sum, col, count, countDistinct, avg

#  Country-wise Sales Analysis
df_country_sales = (
    df_retails_silver.groupBy("country")
    .agg(
        sum(col("total_amount")).alias("total_revenue"),
        count(col("transaction_id")).alias("total_transactions"),
        countDistinct(col("customer_id")).alias("unique_customers"),
        avg(col("total_amount")).alias("avg_transaction_value")
    )
    .orderBy(col("total_revenue").desc())
)

# Save the result as a Gold table
df_country_sales.write.mode('overwrite').saveAsTable('--catalogname--.--schema name--.country_sales_gold')
```

```python
# Product-Store Performance Matrix
df_product_store_matrix = df_retails_silver.groupBy("product_name", "store_name", "location") \
    .agg(
        sum(col("total_amount")).alias("revenue"),
        sum(col("quantity")).alias("quantity_sold")
    ) \
    .orderBy(col("revenue").desc())

df_product_store_matrix.write.mode('overwrite').saveAsTable('--catalogname--.--schema name--.product_store_matrix_gold')
```

```python
# customer_segmentation
from pyspark.sql.functions import sum, col, count, avg, when

df_customer_segmentation = df_retails_silver.groupBy("customer_id", "full_name", "country") \
    .agg(
        sum(col("total_amount")).alias("total_spent"),
        count(col("transaction_id")).alias("purchase_frequency"),
        avg(col("total_amount")).alias("avg_order_value")
    )

df_customer_segmentation = df_customer_segmentation.withColumn(
    "customer_segment",
     when((col("total_spent") > 10000) & (col("purchase_frequency") > 3), "VIP")
    .when((col("total_spent") > 5000) & (col("purchase_frequency") > 2), "High Value")
    .when(col("total_spent") > 2000, "Regular")
    .otherwise("Occasional")
)

df_customer_segmentation.write.mode('overwrite').saveAsTable('--catalogname--.--schema name--.customer_segmentation_gold')
```

```python
# Daily Sales Summary (for time series analysis)
df_daily_sales = df_retails_silver.groupBy("transaction_date") \
    .agg(
        sum(col("total_amount")).alias("daily_revenue"),
        count(col("transaction_id")).alias("daily_transactions"),
        avg(col("total_amount")).alias("avg_transaction_value"),
        sum(col("quantity")).alias("total_items_sold")
    ) \
    .orderBy("transaction_date")

df_daily_sales.write.mode('overwrite').saveAsTable('--catalogname--.--schema name--.daily_sales_gold')
```

```python
# Product Performance with Rankings
windowSpec = Window.orderBy(col("total_revenue").desc())

df_product_rankings = df_retails_silver.groupBy("product_id", "product_name", "category", "price") \
    .agg(
        sum(col("total_amount")).alias("total_revenue"),
        sum(col("quantity")).alias("total_quantity_sold"),
        count(col("transaction_id")).alias("total_transactions")
    ) \
    .withColumn("revenue_rank", dense_rank().over(windowSpec))

df_product_rankings.write.mode('overwrite').saveAsTable('--catalogname--.--schema name--.product_rankings_gold')
```

```python
# Location-Category Analysis (Heat Map Data)
df_location_category = df_retails_silver.groupBy("location", "category") \
    .agg(
        sum(col("total_amount")).alias("revenue"),
        sum(col("quantity")).alias("quantity_sold")
    ) \
    .orderBy("location", col("revenue").desc())

df_location_category.write.mode('overwrite').saveAsTable('--catalogname--.--schema name--.location_category_gold')
```

```python
# customer_lifetime_value
from pyspark.sql.functions import (
    sum,
    count,
    avg,
    min,
    max,
    col
)

df_clv = (
    df_retails_silver.groupBy(
        "customer_id",
        "full_name",
        "email",
        "country",
        "registration_date"
    )
    .agg(
        sum(col("total_amount")).alias("lifetime_value"),
        count(col("transaction_id")).alias("total_orders"),
        avg(col("total_amount")).alias("avg_order_value"),
        min(col("transaction_date")).alias("first_purchase_date"),
        max(col("transaction_date")).alias("last_purchase_date")
    )
    .orderBy(col("lifetime_value").desc())
)

df_clv.write.mode('overwrite').saveAsTable('--catalogname--.--schema name--.customer_lifetime_value_gold')
```

```python
# Store-Country Revenue Distribution
df_store_country = df_retails_silver.groupBy("store_name", "location", "country") \
    .agg(
        sum(col("total_amount")).alias("revenue"),
        countDistinct(col("customer_id")).alias("unique_customers")
    ) \
    .orderBy(col("revenue").desc())

df_store_country.write.mode('overwrite').saveAsTable('--catalogname--.--schema name--.store_country_distribution_gold')
```
